```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd

file_path = r"C:\Users\leube\Ironhack\Bootcamp\week_3\ironhack_gpw3\ironhack_gpw3\data\raw\air_quality_health_assesments.csv"
df = pd.read_csv(file_path)
df
```

```python

import numpy as np

def fill_na_with_group_mean(group):
    return group.fillna(group.median())

df['Premature Deaths'] = df.groupby(['City', 'Year'])['Premature Deaths'].transform(fill_na_with_group_mean)

df['Premature Deaths']
```

```python
df['Year'] = df['Year'].astype(int)  

df_filtered = df[(df['Year'] >= 2017) & (df['Year'] <= 2021)]

df_grouped = df_filtered.groupby('Year')['Premature Deaths'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_grouped['Year'], df_grouped['Premature Deaths'], marker='o', linestyle='-', color='b')
plt.title('Premature Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('Premature Deaths')
plt.grid(True)
plt.xticks(df_grouped['Year'])
plt.show()
```
